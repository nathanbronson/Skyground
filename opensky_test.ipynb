```python
from opensky_api import OpenSkyApi
import numpy as np
from tqdm import tqdm
api = OpenSkyApi()
```

```python
def get_in_progress_paths(box=(15, 70, -135, -65), only_first=None):
    states = api.get_states(bbox=box)
    crafts = set([s.icao24 for s in states.states])
    tracks = [api.get_track_by_aircraft(c) for c in tqdm(list(crafts)[:only_first])]
    paths = [{
        "icao24": t.icao24,
        "callsign": t.callsign,
        "lats": [i[1] for i in t.path],
        "longs": [i[2] for i in t.path],
        "alts": [i[3] for i in t.path]
    } for t in filter(lambda x: x is not None, tracks)]
    return [np.array([p["lats"], p["longs"], p["alts"]]) for p in paths]
```

```python
arr = get_in_progress_paths(only_first=100)
arr
```
